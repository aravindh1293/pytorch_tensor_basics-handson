## tensor reshaping

```
tensor = torch.rand(784); tensor
```

```
viewed_tensor = tensor.view(1,28,28); viewed_tensor
```

```
 viewed_tensor.shape
```

```
reshaped_tensor = tensor.reshape(1,28,28); reshaped_tensor.shape
```

it is always advisable to use `.reshape` instead of `.view`.

## pytorch's Image type operations

```
image = torch.rand(600,400,3)
```

```
image.shape
```

images are generaaly stored in computer as `[height, width, channels]`. but pytorch operates in `[channels, height, width]`

```
pytorch_image = image.permute(2,0,1); pytorch_image.shape
```
